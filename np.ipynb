```python
store_count = df_purchase.groupby('Store')['Store'].count().reset_index(name="Count")
store_count
```

```python
prctg = lambda x: round(x / store_count['Count'].sum() * 100, 1)
```

```python
store_count['Percentage'] = prctg(store_count['Count'])
store_count
```

```python
store_count = store_count.sort_values(by='Percentage', ascending=False)
store_count
```

```python
store_count['% acum'] = store_count['Percentage'].cumsum()
store_count
```

```python
store_count = store_count.reset_index()
store_count.head(20)
```
